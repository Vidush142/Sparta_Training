## Visualisations

```python
copy_df = play_store_df.copy()
```

```python
copy_review_df = play_store_reviews_df.copy()
```

```python
df_merged = pd.merge(copy_df, copy_review_df, on='App')
```

```python
sb.boxplot(y='Rating', x='Sentiment', data=df_merged)
```

```python
sb.barplot(x='Type', y='Sentiment_Polarity', data=df_merged)
```

```python
sb.barplot(y='Category', x='Sentiment_Polarity', data=df_merged, hue='Category')
```

```python
pivot_table = df_merged.pivot_table(index='Category', values='Sentiment_Polarity', aggfunc='mean')
```

```python
sb.heatmap(pivot_table,cmap='coolwarm', annot=True, linewidths=0.5)
```

```python
sb.countplot(x='Type', data=df_merged)
```

```python
sb.lineplot(x='Price', y='Rating', data=df_merged)
```

```python
sb.scatterplot(x='Rating', y='Installs', data=df_merged)
```

```python
sb.regplot(x='Rating', y='Installs', data=df_merged, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
```

```python
def categorize_rating(rating):
    if rating < 2.5:
        return 'Poor'
    elif rating < 3.5:
        return 'Average'
    elif rating < 4.5:
        return 'Good'
    else:
        return 'Excellent'
```

```python
df_merged['rating_range'] = df_merged['Rating'].apply(categorize_rating)
```
